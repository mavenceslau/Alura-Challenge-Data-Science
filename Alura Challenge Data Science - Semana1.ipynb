#### Dicionário de dados

* `customerID`: número de identificação único de cada cliente
* `Churn`: se o cliente deixou ou não a empresa 
* `gender`: gênero (masculino e feminino) 
* `SeniorCitizen`: informação sobre um cliente ter ou não idade igual ou maior que 65 anos 
* `Partner`:  se o cliente possui ou não um parceiro ou parceira
* `Dependents`: se o cliente possui ou não dependentes
* `tenure`:  meses de contrato do cliente
* `PhoneService`: assinatura de serviço telefônico 
* `MultipleLines`: assisnatura de mais de uma linha de telefone 
* `InternetService`: assinatura de um provedor internet 
* `OnlineSecurity`: assinatura adicional de segurança online 
* `OnlineBackup`: assinatura adicional de backup online 
* `DeviceProtection`: assinatura adicional de proteção no dispositivo 
* `TechSupport`: assinatura adicional de suporte técnico, menos tempo de espera
* `StreamingTV`: assinatura de TV a cabo 
* `StreamingMovies`: assinatura de streaming de filmes 
* `Contract`: tipo de contrato
* `PaperlessBilling`: se o cliente prefere receber online a fatura
* `PaymentMethod`: forma de pagamento
* `Charges.Monthly`: total de todos os serviços do cliente por mês
* `Charges.Total`: total gasto pelo cliente

```python
#Importando Bibliotecas

import pandas as pd
import numpy as np
```

```python
#Configurações de exibição

pd.set_option("display.max_columns", 100)
pd.set_option("display.max_colwidth", None)
```

```python
# Importando os dados

data = pd.read_json(
    path_or_buf = 'dataset/Telco-Customer-Churn.json',
    orient='columns')
data.head(10)
```

Extração de dados das colunas:

```python
# Normalize customer

data_customer = pd.json_normalize(data=data.customer)
data_customer.head(10)
```

```python
# Normalize phone

data_phone = pd.json_normalize(data=data.phone)
data_phone.head(10)
```

```python
# Normalize internet

data_internet = pd.json_normalize(data=data.internet)
data_internet.head(10)
```

```python
# Normalize account

data_account = pd.json_normalize(data=data.account)
data_account.head(10)
```

```python
# Concatenado Dataframe principal com os normalizados

data_final = pd.concat([data, data_customer], axis='columns')
data_final = pd.concat([data_final, data_phone], axis='columns')
data_final = pd.concat([data_final, data_internet], axis='columns')
data_final = pd.concat([data_final, data_account], axis='columns')

# Removendo as colunas que não serão utilizadas

data_final.drop(columns='customer', inplace=True)
data_final.drop(columns='phone', inplace=True)
data_final.drop(columns='internet', inplace=True)
data_final.drop(columns='account', inplace=True)

data_final.head(5)
```

Tratamento de dados

```python
# Tradução do dataset

data_final.columns = ['ClienteId', 'ClienteCancelado', 'Gênero', 'Idoso', 'Parceiro', 'Dependentes', 'TempoContrato', 'ServiçoTelefone', 'MultiplasLinhas',
                      'ServiçoInternet', 'SegurançaOnline', 'BackupOnline', 'ProteçãoDispositivo', 'SuporteTecnico', 'StreamingTV', 'StreamingFilmes',
                      'Contrato', 'FaturaSemPapel', 'FormaPagamento', 'CobrançaMensal', 'CobrançaTotal']
```

```python
# Podemos corrigir os valores vazios da coluna 'CobrançaTotal' por 0

data_final.CobrançaTotal = data_final.CobrançaTotal.replace(' ', '0')
```

```python
# Após algumas sugestões, percebi que o ideal seria manter a coluna 'CobrançaTotal' com valores iguais a coluna 'CobrançaMensal'
# nos casos que clientes não completaram 1 mes de contrato

cobrancaTotal = lambda data: data['CobrançaMensal'] if data['CobrançaTotal'] == 0 else data['CobrançaTotal']
data_final.CobrançaTotal = data_final.apply(cobrancaTotal, axis = 1)
```

```python
# Conferencia da substituição de valores

data_final.query("TempoContrato == 0")
```

```python
# Tipo de Dados

data_final.info()
```

```python
# A coluna 'CobrançaTotal' está como 'object' no tipo de dados, mas por se tratar de valores, a mesma será convertida em 'float'

data_final['CobrançaTotal'] = pd.to_numeric(data_final['CobrançaTotal'], errors='coerce')
data_final.info() 
```

```python
# os valores vazios da coluna 'ClienteCancelado' não interessa, uma vez que queremos saber a taxa de cancelamento 

data_final = data_final.drop(data_final[data_final.ClienteCancelado == ''].index, axis=0) 
data_final.info()
```

```python
# A notação Boleana da da coluna 'Idoso' está em um padrão diferente das demais, logo foi feita a conversão desses valores

data_final.Idoso.value_counts()
```

```python
# Feita a alteração, onde o '0' foi substituido por 'No', e '1' por 'Yes'

for idoso in 'Idoso':
    data_final['Idoso'] = data_final['Idoso'].replace({0: "No", 1: "Yes"})

data_final.Idoso.head(10)
```

```python
# Criando coluna Contas diárias
# Os gastos diários serão a divisão dos valores da coluna 'Cobrança_Mensal' divididos por 30 dias

data_final_contas_diarias = round(data_final['CobrançaMensal'] / 30, 2)

data_final.insert(loc = 19, column = 'ContasDiarias', value = data_final_contas_diarias) 

data_final.head(10)
```

```python
# Exportando os Dados para um arquivo .csv

data_final.to_csv(path_or_buf = 'dataset/Telco-Customer-Churn-Tratado.csv', index = False)
```
